```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip((1, 3, 9), axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(f"{n_neighbors} neighbor/s")
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
axes[0].legend(loc='lower left')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42
)

training_acc = []
test_acc = []

neighbors_settings = range(1, 12)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_acc.append(clf.score(X_train, y_train))
    test_acc.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_acc, label='train set accuracy')
plt.plot(neighbors_settings, test_acc, label='test set accuracy')
plt.ylabel('Accuracy')
plt.xlabel('n_neighbors')
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)

print('Test set predictions:\n{}'.format(reg.predict(X_test)))
```

```python
# 1 - perfect prediction, 0 - train set mean. The score is good only for some seeds(definitely not 42 this time)
print('Test set R^2: {:.2f}'.format(reg.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip((1, 3, 9), axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
        f'{n_neighbors} neighbors\n train score: {reg.score(X_train, y_train):.2f}, test score: {reg.score(X_test, y_test):.2f}'
    )
    ax.set_xlabel('Feature')
    ax.set_ylabel('Target')
axes[0].legend(['Model prediction', 'Train data/score', 'Test data/score'], loc='best')
```

# Linear models

regression: y = w[0] * x[0] + w[1] * x[1] + [...] + w[p] * x[p] + b

p - no. of data features

w, b - model parameters

y - model prediction

single feature:
y = w[0] * x[0] + b (ax + b line)

# Controlling decision tree complexity

```python
from sklearn.tree import DecisionTreeClassifier
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42
)
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
print('Train set accuracy: {:.3f}%'.format(100*tree.score(X_train, y_train)))
print('Test set score: {:.3f}%'.format(100*tree.score(X_test, y_test)))
```

```python
cut_tree = DecisionTreeClassifier(random_state=42, max_depth=4)
cut_tree.fit(X_train, y_train)
print('Train set accuracy {:.3f}%'.format(100*cut_tree.score(X_train, y_train)))
print('Test set accuracy: {:.3f}%'.format(100*cut_tree.score(X_test, y_test)))
```

Decision tree analysis

```python
from sklearn.tree import export_graphviz
export_graphviz(cut_tree, out_file='tree.dot', class_names=['malignant', 'benign'],
                feature_names=cancer.feature_names, impurity=False, filled=True)

with open('tree.dot') as f:
    dotgraph = f.read()
graphviz.Source(dotgraph)
```

```python
#Regressors based on decision trees cannot predict outside train set data
plt.figure(figsize=(15,10))
import os
ram_prices = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH, "ram_price.csv"))

plt.semilogy(ram_prices.date, ram_prices.price)
plt.xlabel('Year')
plt.ylabel('Price in $/MB')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
data_train = ram_prices[ram_prices.date < 2000]
data_test = ram_prices[ram_prices.date >= 2000]
#predict prices based on date
X_train = data_train.date[:, np.newaxis]
#logarithmic transform to get simpler data:target relationship
y_train = np.log(data_train.price)
tree = DecisionTreeRegressor(max_depth=4, random_state=42).fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)
#base prediction on all data
X_all = ram_prices.date[:, np.newaxis]
pred_tree = tree.predict(X_all)
pred_lr = linear_reg.predict(X_all)
#revert log transform
price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)
```

```python
#n_estimators - number of trees in a forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=66)
forest = RandomForestClassifier(random_state=42, n_estimators=5)
forest.fit(X_train, y_train)
print('Train set accuracy: {:.3f}%'.format(100*forest.score(X_train, y_train)))
print('Test set accuracy: {:.3f}%'.format(100*forest.score(X_test, y_test)))
#trees are stored in estimator_ parameter
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20,10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title(f'Tree {i}')
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1], alpha=0.4)
    axes[-1, -1].set_title('Random forest')
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_blobs, make_circles

X, y = make_circles(noise=0.25, factor=0.5, random_state=1)
y_named = np.array(['blue', 'orange'])[y]

X_train, X_test, y_train_named, y_test_named, y_train, y_test = train_test_split(
    X, y_named, y, random_state=42
)

gbrt = GradientBoostingClassifier(random_state=42).fit(X_train, y_train_named)
```

The Decision function

```python
#The value returned is of (n_samples, ) shape with float value for every sample
print(f'X_test.shape: {X_test.shape}')
print(f'Decision function shape: {gbrt.decision_function(X_test).shape}')
```

```python
#First few Decision values (positive means 'positive' class, negative means 'negative' class for binary data)
print(f'Decision function:\n{gbrt.decision_function(X_test)[:6]}')
```

```python
#We can also return boolean values
print(f'Decision function results:\n{gbrt.decision_function(X_test) > 0}')
print(f'Prediction:\n{gbrt.predict(X_test)}')
```

```python
#classes_[0] - negative class, classes_[1] - positive class
#convert booleans to integer values
positive = (gbrt.decision_function(X_test) > 0).astype(int)
#same as gbrt.prediction
pred = gbrt.classes_[positive]
print(f'Is pred == predicitons? {np.all(pred == gbrt.predict(X_test))}')
```
